```python
#Making 5 classes of countries with least happy, normal, not happy, happy, very happy, most happy for 2010
```

```python
def binsfunct(value):
    
    return ('Least Happy' if (value >130) 
            else 'Not Happy' if(110 <= value <= 130) 
            else 'Normal' if (80 <= value <= 110) 
            else 'Happy' if (50 <= value <= 80)
            else 'Very Happy' if (20 <= value <= 50)
            else 'Extremely Happy')
            

merged['HappinessClass_2010'] = merged['HappinessRank_2010'].map(binsfunct)
merged['HappinessClass_2016'] = merged['HappinessRank_2016'].map(binsfunct)
merged['HappinessClass_2017'] = merged['HappinessRank_2017'].map(binsfunct)
```
